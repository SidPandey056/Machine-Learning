## Random Variable 
- is a set of possible values from a random experiment.
- For example; in a coin toss random variable $X = \{0,1\}$ where $0$ is Tail and $1$ is Head
- Here, $X$ is a **Random Variable** and $\{0,1\}$ is **Sample Space**

## Types of Random Variables:
1. Discrete Random Variables
- Any random variable holding discrete values is a discrete random variable.
- For example, sample space of a die roll $X = \{1,2,3,4,5,6\}$, coin toss $X = \{0,1\}$

2. Continuous Random Variable
- Any random varibale holding a continuous values is a continuous random variable
- For example, height of students in a class (i.e can range between number, say $5\ ft.$ to $6.5\ ft.$)
- GPA of students in a class

# Probability Distribution
- Probability distribution is a list sample space and their associated probabilities 
- For example; probability distribution table of a coin toss is given as:

```python
pd.DataFrame({'Sample Space': ['H', 'T'], 'Probability': ['1/2','1/2']})
```

- Probability Distribution for a die roll

```python
pd.DataFrame({'Sample Space': [1,2,3,4,5,6], 'Probability': ['1/6','1/6','1/6','1/6','1/6','1/6']})
```

## Problem with Probability Distribution
- As of discrete random variables with small sample sapce, the probability distribution can be easily shown in a table
- If there is large sample sample sapce within discrete random variable or we're working with continuous random variables, it is tedious or almost impossible

## Solution:
- As a solution to such problem, we model the realtionship between the sample sapce and probability with a mathematical function
- And the function is called as **Probability Distribution Function** 
- The function then can be plotted in a graph

# Types of Probability Distribution Function
1. Probability Mass Function (PMF)
- PMF describes the distribution function of a discrete random variable
- The PMF of a discrete random variable assigns probability to each sample space of a random variable
- The probabilities assigned by the **PMF must satisfy two conditions**:
    - The probability assigned to each value must be non-negative (i.e greater than or equal to zero)
    - The sum of probabilities assigned to each value must equal 1

2. Probability Density Function (PDF)
- PDF is a mathematical function that describes the probability distribution of a continuous random variable

3. Cumulative Distribution Function (CDF)
- CDF says the probability at $x$ is the probability of outcome at a value less than or equal to x
$$Mathematically,\ F(x) = P(X\leq x) $$

- See it as, the CDF at x is the sum of probabilities upto the point x (including x)

# Random Experiment for die roll

```python
import numpy as np
```

```python
L = []
for x in range(10000):
    x = np.random.randint(1,7)
    L.append(x)
```

```python
outcome = pd.Series(L).value_counts()
```

```python
a = (outcome/outcome.sum()).sort_index()
a
# The value of 'a' is 1/6 for each sample space
# Since it is random experiment it is not exactly 1/6 but is around 1/6
```

```python
a.plot(kind = 'bar')
```

# Probability of sum of outcome two dice roll

```python
L = []
for x in range(10000):
    a = np.random.randint(1,7)
    b = np.random.randint(1,7)
    L.append(a+b)
```

```python
outcome = pd.Series(L).value_counts().sort_index()
outcome = outcome / outcome.sum()
outcome
```

```python
outcome.plot(kind = 'bar')
```

```python
# CDF plot
cum_sum = np.cumsum(outcome)
cum_sum.plot(kind = 'bar')
```
