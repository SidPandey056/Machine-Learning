# Sample Distribution
- A probability distribution of a statistics that is obtained from repeated sampling of a specific population.
- For example: If we take a sample size of 50, a 100 times from a population and calculate mean (could be other statisitcs) every time. The distribution formed by those 100 means is called as sample distribution  
- If we calculate mean of multiple samples, it is normally distributed (According to CLT)
- If we calculate variance of multiple samples (while population is normally distributed with degree of freedom $(n-1)$) it will follow chi-square distribution
# Point Estimate:
- Point estimate is a single numeric value calculated to estimate population parameters from a sample distribution

# Confidence Level
- It indicated how sure we are that the population parameter estimated from sample falls under ther confidence interval. It is expressed in percentage, usually 95%

# Confidence Interval
- A range of values within which we expect a particular population parameter to fall under.
$$ Confidence\ interval = Point\ Estimate\ \pm Margin\ of\ Error $$

> The `parameter` and `statistics` are the terms that describes the population and sample population respectively.
> Mean ($\mu$), standard deviation ($\sigma$) etc. that are represented in greek letters are the parameters and describes the population while the alphabetic characters like mean ($\bar x$), standard deviation ($s$) etc represent sample population 

# Calculation of Confidence Interval
## 1. Z-Procedure (Sigma Known)
- ### Conditions required for z-procedure:
    - Random Sampling
    - Known population standard deviation ($\sigma$)
    - For sample size smaller than 30, the population should be noramlly distributed, if greater than 30 CLT is applied

$$ C.I = \bar X \pm Z_{\alpha / 2} \frac{\sigma}{\sqrt n} $$
- Here, $Z$ is the standard normal distribution whose mean is 0 and standard deviation is 1
- Formula for Z is given as:
$$Z = \frac{\bar X - \mu}{\frac{\sigma}{\sqrt n}} $$

- Confidence level is represented as in percentage as $95\%$  i.e $(1-\alpha) = 95 \%$

> NOTE: A confidence level means that if we were to draw 100 samples from a population and calculate confidence interval for each sample approximately 95 intervals will contain true population parameter 

> NOTE: Higher confidence level implies higher confidence interval. $100\%$ confidence level will have interval $[-\infty, +\infty]$

## 2. T-Procedure
- Conditions required:
    - Random Sampling
    - Sample Standard Deviation
    - Approximate normal ditribution of population, if not use CLT
    - Independent observations
    
$$ C.I = \bar X \pm T_{\alpha / 2} \frac{s}{\sqrt n} $$

$$ T =  \frac{\bar X - \mu}{\frac{s}{\sqrt n}}$$

- Unlike in Z-procedure, we use sample standard deviation ($s$) to calculate the T-distribution.
- It is not reasonable in the Z-procedure that we will have standard deviation of the population. Thus T-distribution was introduced.
- The T-distribution is known as Student's T distribution.
- The only parameter in t-distribution is degree of freedom i.e $ df = n-1$ where $n$ is the sample size

### Intuition:
- The sample standard deviation $s$ will differ with random samples
- t-distribution is similar to the z distribution except for the fact that the lower degree of freedom has flatter tails
- and for $df= \infty$, t-distribution is same as z-distribution
- The t-procedure will have have high range confidence intervals for lower degree of freedom.
