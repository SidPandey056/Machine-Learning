# Crossvalidation
- Crossvalidation is a process of splitting dataset into train and test data into multiple sets or folds
- In normal `train_test_split` the data set is splitted into single train and test data
#### - In crossvalidation with K-fold validation (example: k = 5 folds): 
- the dataset is divided into 5 folds where in one iteration single fold is used as validation set and other are used as train set
- while fold 1 is validation set, fold 2,3,4,5 are used as train set
- while fold 2 is validation set, fold 1,3,4,5 are used as train set
- This goes until all folds are used as validation set
- if k = 5, the number of errors we get to analyze is 5.
- From these errors we calculate the average error and choose a model with least error.



> This will increase the model's efficiency and evaluation as it is trained on all the possible data points and at the same time is evaluated on the same data set without leaking information between folds.
## Types of Crossvalidation
1. <b>K-Fold Cross-Validation</b>:
    - Divides the data into K folds (e.g., 5-fold or 10-fold) and performs K training and testing iterations.
    - Larger the value of k expensive the computation
    - When k equals to number of rows, it is Leave-One-Out crossvalidation

3. <b>Stratified K-Fold</b>: Ensures each fold maintains the proportion of classes (for classification tasks with imbalanced data).
4. <b>Leave-One-Out Cross-Validation (LOO-CV)</b>: Each test set contains only one data point, while the rest of the data is used for training. This is very time-consuming but useful for small datasets.
