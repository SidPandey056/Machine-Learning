# Classification Metric
- It is likely that our model will not perform perfectly.
- It means there are 4 possibilities to a classification's outcome:
1. Predicts true for actual true (true positive)
2. Predicts false for actual false (true negative)
3. Predicts false for actual true (false positive)
4. Predicts true for actual false. (false negative)

- 1 and 2 are the outcome we want to achieve
- But the model trained are not perfect thus can result some predictions to be incorrect as 3 and 4

## Imbalance classes
- Imbalance class are where a number of classes holds large amount of data than others
- For example, in a dataset of credit card transaction, fraudulent transaction (1) and legitimate transaction (0)
- The fraudulent cases are generally rare causing the imbalance class in the data set where most of the data are labeled 0 and few labeled 1
- In such case it is often to have accuracy paradox, thus we should not only rely on accuracy matric.
> Espically while dealing with imabalanced classes, it is not appropriate to solely reply on accuracy matric.


## Recall
- Recall is a performance metric for a classification model
- The formuTrela to calculate Reacll is:
$$ Recall = \frac{True\ Positive}{Total\ Actual\ Positive}$$
- From the confusion matrix, we can see that, $TP = 4$ and $Total AP = (4+1) = 5$
$$\therefore Recall = \frac{4}{5} = 0.8$$
> Recall describes the relavent cases, for the example it shows 80% of the time we found the relavent case. The model is able to predict infected for infected persons 80% of the time.

- From the image in accuracy paradox, when the model predicts all healthy the recall is given as:
$$ Recall = \frac{TP}{Total\ Actual\ Positive} = \frac{0}{5} = 0 $$
- It is when the model alerts that it is not catching any relavent cases and is overfitted and reports back everything to the majority class. 

## Precision
- When the prediction is Positive, how often it is correct
- It is calculated as:
$$ Precision = \frac{True\ Positive}{Total\ Predicted\ Positive} $$
- For the case :

- Here, $ Precision = \frac{4}{6} = 0.666$

- For the case, an incorrect model predicting majority class (see accuracy paradox), $ Precision = \frac{0}{0}$ i.e undefined

> Based on model we're using there is tradeoff between the precision and recall, which will be learnt later in ROC (Receiver Operator Characteristics)

## F1 score (or F metric or F score)
- F1 score is the harmonic mean of the recall and precision metric
$$F = \frac{2 \times precision \times recall}{precision + recall}$$

- Instead of normal mean we use harmonic mean beacuse, it allows the F score to go to zero if one of recall or precision is zero.

> If we see F score is zero, either recall or precision or both are zero

## Trade off between TP and FP
- There can be trade-off between the true positive and false positive
- For cases like virus detection, it would be appropriate to minimize false negative
- For example; we would not like to predict a virus tested negative while the person is actually infected
- Infact it is better to predict tested positive even if one is not infected with virus.
- For such cases we can trade-off between TP and FP by decreasing the cut-off point

> Imagine in a logistic curve with cut-off point at 0.5,
> - TP (above cut-off) = 4
> - TN (below cut-off) = 3
> - FP (should be below but above cut-off) = 1
> - FN (should be above but below cut-off) = 2 (say above 0.2)
> - As described above we don't want false negative, so the cut-off point should be set to 0.2 to minimize false negative



## AUC (Aread under the curve)
- The best possible area under the curve for ROC curve is 1 (for TPR vs FPR)
- We can compare between the models with AUC. The more the AUC the more best performing model.
- We can also create precision vs recall curve and compare between models

## Intuition
- For the intuition analyze the graph above and see the formula
- The recall is rate of true positive over actual positive
- The precision is rate of true positive over predicted positive
- As threshold increases the number of false negative increases (and false positive decreases) , hence the recall is low (and precision is high) 
- As threshold decreases the number of false negative decreases (and false positive increases), hence the recall is high (and precision is low)

- At threshold zero, the recall is 1 and precision is zero
- At threshold one, the recall is zero and precision is 1.
