# Hyperplane
- In an N-dimensional space, a hyperplane is a flat subspace of hyperplane dimension $N-1$

<u>**Example**</u>:
- 1-D Hyperplane is a single point
- 2-D Hyperplane is a line
- 3-D Hyperplane is a flat plane

> The main idea bhind SVM is that we use hyperplanes to create a separation between classes.

## Support Vector Classifier (Soft Margins)
- When there is a perfectly separable classes in a data set maximal margin classifier is fine
- but when some data points are more aligned to opposing class we need soft margin to allow some bias in the model.
- Soft margin allows some points to be misclassified.
- Rather using **_Maximal Margin Classifier_** to tightly separate classes, we use **_Support Vector Classifier_** that allows some bias.

## Kernel and Kernel Trick
### Kernel:
- Kernel allows us to project a original features to a higher dimension where a support vector classifier is unable to effectively classify the data points.
- This kernel performs explicit transformation of data points to a higher dimension and the dot product of the features is calculated
### Kernel Trick
- Unlike kernels, the kernel trick does not perform explicit computation for the higher order dimension rather it directly applies kernel function to the original dataset.
- While kernel are very much computationally expensive for higher dimension calculations, Kernel trick does the same with comparatively less computation

> Kernels gives a measure of similarity between the original feature space and enlarged feature space, and use kernel trick to perform calculations instead of saving a lot of computations
