# Train_Validation_Split Test

```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Advertising.csv').drop('Unnamed: 0', axis = 1)
```

```python
df.head()
```

```python
X = df.drop('sales',axis = 1)
y = df['sales']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, root_mean_squared_error
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
```

```python
# train_validation_test split
# split the train_test as usual but the test(here 'other') will be further splitted into eval and test data
X_train, X_other, y_train, y_other = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_eval, X_test, y_eval, y_test = train_test_split(X_other, y_other, test_size = 0.5)
```

```python
# feature scaling
scaler = StandardScaler()
scaler.fit(X_train)
scaled_X_train = scaler.transform(X_train)
scaled_X_eval = scaler.transform(X_eval)
scaled_X_test = scaler.transform(X_test)
```

```python
# Model 1
model1 = Ridge(alpha = 20)
```

```python
model1.fit(scaled_X_train, y_train)
```

```python
y_eval_pred = model1.predict(scaled_X_eval)
```

```python
# MSE of model 1
mean_squared_error(y_eval, y_eval_pred)
```

```python
# Model 2
model2 = Ridge(alpha= 1)
```

```python
model2.fit(scaled_X_train, y_train)
```

```python
y_eval_pred_2 = model2.predict(scaled_X_eval)
```

```python
# MSE of model 2
mean_squared_error(y_eval, y_eval_pred_2)
```

#### After tuining hyperparameter with the validation model, we can finally evaluate the best performing model (i.e model2) and report/publish the model's performance during deployment

```python
y_test_pred = model2.predict(scaled_X_test)
```

```python
mean_squared_error(y_test, y_test_pred)
```

```python
root_mean_squared_error(y_test, y_test_pred)
```

> the metrics of eval and test and test will be different after the kernel restart, this is because the random_state is not selected for validation_test split
