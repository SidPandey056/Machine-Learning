# KNN (K-Nearest Neighbour)
- KNN is a basic classifier that uses proximity to predict a class of an individual dataset
- The value of $K$ decides the number of data points in proximity to predidct a class
- Even value of $K$ might encounter a condition where tie may occur between prediction class.
- To avoid such cases, choose an odd value or reduce the value of $k$ by $1$ until tie breaks 
- In the image below, for $k=4$ it will create a tie between the classes


- At the same condition choosing $k=5$ will predict the datapoint to be in square class
- The test-datapoint has 3 squares and 2 triangles that are in proximity. Thus greater number of squares are in proximity, predicts the square class.
>NOTE : In case of tie, the scikit-learn will choose the class that appears first. And the ordering is done on the basis of distance. That would be, if tie, see k = 1

## Choosing a value of K
- The objective is to choose a value of $k$ such that error is optimal
$$ Error = 1 - Accuracy$$
- There are two methods such that we get optimal error
  
    1. Elbow method
    2. Crossvalidate a grid search of multiple $k$ values and choose $k$ that results in minimim error or maximum accuracy

1. <u>**Elbow Method**:
- Plot an error rate vs k-value graph and analyze a point where increasing the value of $k$ doesnot subsequently decrease reasonale error.
- In the figure, we can notice that k-value greater than 10/15 doesnot descrease any resonale error rate.
- Although at around $k=22$ the error is minimum but it's only matter of $0.01\%$ which might not be reasonable to increase model complexity for such small amount of error rate.


2. <u>**Crossvalidate with grid search (recommended)**:
- Rather than using elbow method to manually look for best value of $k$
- It is better to make a grid search for multiple value of $k$.
- The `GridSearchCV` of sklearn will generalize the model for best value of $k$ than we human can analyze for it manually.

## KNN Algorithm
- Choose a value of $k$
- Sort feature vector (N dimensional space) by a distance metric
- Choose class based on $K$ nearest feature vector

## KNN Considerations: 
1. <u>**Distance metrics**:
- Minkowski
- Euclidean
- Manhattan
- Chebyshev

2. <u>**Scaling is necessary**</u>: Data points across features may have unnecessary distance, generalizing them would help in model performance
