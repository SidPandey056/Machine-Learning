# Errors and their Role in ML
- Model : $$ \hat{y} = \beta_0x_0 + \beta_1x_1 + ..... + \beta_nx_n   $$
$$Or$$
$$ \hat{y} = \sum_{i=0}^n \beta_ix_i $$
- There is usually no set of $\beta$ that creates perfect fit to $y$

- What is residual error:
$$ e = y_i - \hat y_i$$
- The Sum of Squared Residuals is called as Least Square error.
- what we've know as cost function is the average of SSR.
- Sum of Squared Residuals (SSR) or Least Square Error (LSE) is given as:
$$ SSR/LSE = \sum_{i=0}^{m-1} e^2 =\sum_{i=0}^{m-1}(y_i - \hat y_i)^2$$
- Mean Absolute Error (MAE):
$$ MAE = \frac{1}{m} \sum_{i=0}^{m-1} |y_i - \hat y_i|$$
- Mean Square Error (MSE):
$$ MSE = \frac{1}{m} \sum_{i=0}^{m-1} e^2 = \frac{1}{m} \sum_{i=0}^{m-1}(y_i - \hat y_i)^2$$
- Root Mean Square Error (RMSE) : is used to report back in the original unit as of the $y$'s unit
$$ RMSE = \sqrt{\frac{1}{m} \sum_{i=0}^{m-1}(y_i - \hat y_i)^2}$$

> What value of RMSE is appropriate, it depends in the context, for example: if we're predicting price of a house RMSE of 10\$ is fantasic but horrible for predicting the price of a candy bar. But in health care and medical predictions even small RMSE may have significant impact.

> NOTE: When the training MSE is smaller but the test MSE is larger, the model is said to be overfitted.
- It a fundamental of Machine learning that we fit the model such that it will have less MSE for the unseen data. This way true property of a function $f$ is known.
- It is obvious to have to have less residuals with the training dataset.
- The challenge is to identify the model that will best fit for the unseen data

## <u>Random error or irreducible error:</u>
- In the model there is alway random error that is occured because of the data points that are not included in the training set such error is denoted by $\epsilon$ (epsilon). $$ \hat y_i = f(X_i) + \epsilon_i $$



# Simple Linear Rergession
> NOTE : Simple linear regression has only one dependent $(y)$ and one independent variable $(x)$
- The regression coefficients $\beta_0$ and $\beta_1$ can be calcuated using `Pearson Correlation Coefficient` and `Standard Deviation` as
$$ \beta_1 = r_{x,y}. \left( \frac{\sigma_y}{\sigma_x} \right)$$
$$\beta_0 = \bar y - \beta_1 \bar x $$
    - $r$ is Pearson Correlation Coefficient
    - $\sigma_x\ and\ \sigma_y$ are Standard Deviations
    - $\bar x\ and\ \bar y$ are the mean
